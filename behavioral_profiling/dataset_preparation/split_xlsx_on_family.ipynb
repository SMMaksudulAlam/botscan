```python
import pandas as pd

# Input file (CSV now, not Excel)
input_file = "c2_data_all_04_29_25.csv"
output_file = "splited_data_on_family.xlsx"

# Load CSV
df = pd.read_csv(input_file, encoding="latin1", low_memory=False)

# Normalize columns
df["malware"] = df["malware"].astype(str).str.lower()
df["malware_printable"] = df["malware_printable"].astype(str).str.lower()

# Family definitions
families = {
    "mirai": ["mirai"],
    "gafgyt": ["gafgyt", "bashlite"],
    "mozi": ["mozi"],
    "hajime": ["hajime"],
    "kaiji": ["kaiji"]
}

with pd.ExcelWriter(output_file, engine="xlsxwriter") as writer:
    
    for family, keywords in families.items():
        mask = df["malware"].apply(lambda x: any(k in x for k in keywords)) | \
               df["malware_printable"].apply(lambda x: any(k in x for k in keywords))
        
        filtered_df = df[mask]

        print(f"{family}: {len(filtered_df)} rows")

        filtered_df.to_excel(writer, sheet_name=family, index=False)

print(f"✅ Output saved to {output_file}")
```
